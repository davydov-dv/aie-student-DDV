Arange создаёт массив по правилу: если параметр 1 число, то создает массив от 0 до нашего числа
не включительно с шагом 1
если параметров 2, то создает массив от первого числа до второго не включительно с шагом 1
если параметром 3, то создает массив в промежутке от одного числа до другого с шагом указанным
3 параметром. От какого числа до какого зависит от знака 3 параметра. Если он положительный, то
от первого параметра до второго не включительно.
Если он отрицательный, то от второго до первого не включительно

```python
arr_arange_pos = np.arange(10, 11, 0.1)
arr_arange_pos
```

```python
arr_arange_neg = np.arange(11, 10, -0.0001)
arr_arange_neg
```

Примеры векторных операций и базовых статистик:

поэлементное сложение и умножение массивов

```python
first_arr = np.arange(10)
first_arr
```

Ниже каждый элемент в массиве поэлементно возводится в квадрат

```python
second_arr = np.arange(10) ** 2
second_arr
```

Поэлементное сложение массивов (значение с индексом 0 первого массива складывается с индексом 0 второго массива
значение с индексом 1 первого массива складывается с индексом 1 второго массива и так далее. В результате получается новый массив - результат сложения двух массивов)

```python
result_add = first_arr + second_arr
result_add
```

Аналогично сложению, происходит только смена операции

```python
result_sub = first_arr - second_arr
result_sub
```

Анологтчно сложению, происходит только смена операции

```python
result_mult = first_arr * second_arr
result_mult
```

Готовый метод в numpy, который сразу находит сумму массива

```python
np.sum(result_add)
```

Готовый метод в numpy, который сразу находит среднее значение в массиве

```python
np.mean(result_add)
```
