Функция для автоматического поиска отрицательных значений. Возвращает индексы, где обнаружились отрицательные значения

```python
def neg_index(data, col):
    isNeg = data[col] < 0
    count = -1
    neg_index = []
    for i in isNeg:
        count +=1
        if(i == True):
            neg_index.append(count)
    return neg_index

neg_user_id_index= neg_index(data, 'user_id')
neg_age_index = neg_index(data, 'age')
neg_purchases_index = neg_index(data, 'purchases')
neg_revenue_index = neg_index(data, 'revenue')
```

<h1>Визуализация

```python
fig, ax = plt.subplots(
    figsize =(10, 5)
)

ax.hist(
    data['revenue'],
    label='выручка',
    bins = 15,
    color = 'lightblue',
    edgecolor= 'white',
    alpha = 0.5
)
ax.set_title('Распределение revenue (выручки) по покупкам (purchases)')
ax.set_xlabel('Выручка')
ax.set_ylabel('покупки')
ax.grid(axis= 'y', linestyle = '--')

fig.savefig('figures/hist_RevPur')
plt.show()
```

График гистограммы показывается, что при снижении числа покупок повышается выручка, что свидетельствует о качестве товара. По сути график говорит, что когда много покупок, то они мелкие (например: ручки, тетрадки или что-то такое). А когда их мало, то они более крупные (например: посудомойка, стулья или что-то такое)

```python
fig, ax = plt.subplots(
    figsize =(10, 5)
)

data.boxplot(
    column = 'purchases', by = 'country', ax = ax,
    patch_artist = True,
    boxprops = dict(facecolor = 'green', color = 'darkgreen'),
    medianprops = dict(color = 'orange', linewidth = 2)
)
plt.suptitle('')

ax.set_title('Количество покупок по странам')
ax.set_xlabel('Страна')
ax.set_ylabel('Покупки')

fig.savefig('figures/boxplot_purchases_country')
plt.show()
```

На графике boxplot видно, что у стран Германия, Франция и Россия одинаковая медиана. Самый стабильный разброс покупок в США, но и самая маленькая медиана. 

```python
fig, ax = plt.subplots(
    figsize =(10, 5)
)

data.boxplot(
    column = 'revenue', by = 'country', ax = ax,
    patch_artist = True,
    boxprops = dict(facecolor = 'purple', color = 'darkred'),
    medianprops = dict(color = 'orange', linewidth = 2)
)
plt.suptitle('')

ax.set_title('Выручка по странам')
ax.set_xlabel('Страна')
ax.set_ylabel('Выручка')

fig.savefig('figures/boxplot_revenue_country')
plt.show()
```

На графике видно, что самой высокой средней вырочкой обладает Германия (Китай близок по значению), а самой маленькой средней вырочкой обладает США (также Франция и Россия близки по значению), но Россия обладает самым большим диапозоном по выручке.
